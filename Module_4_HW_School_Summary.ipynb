## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="outer", on=["school_name", "school_name"])
```

```python
#group by schools
school_names = school_data_complete.set_index("school_name").groupby(["school_name"])
school_names.count().head()
```

```python
# group by school type
school_type = school_data.set_index("school_name")["type"]
school_type.count()
```

```python
# total students per school
students_per_school = school_names['Student ID'].count()
students_per_school
```

```python
# school budget
school_budget = school_data.set_index("school_name")["budget"]
school_budget
```

```python
# budget per student
budget_per_student = school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"]    
budget_per_student
```

```python
#average math and reading scores  
average_math_score = school_names["math_score"].mean()
average_read_score = school_names["reading_score"].mean()
```

```python
# passing percentages
passing_math_percent= school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school
passing_read_percent = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school
overall_passing_percent = (passing_math_percent + passing_read_percent)/2
overall_passing_percent
```

```python
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Student Population": students_per_school,
    "School Budget": school_budget,
    "Budget Per Student": budget_per_student,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_read_score,
    "% Passing Math": passing_math_percent,
    "% Passing Reading": passing_read_percent,
    "Overall Passing Rate": overall_passing_percent
})

school_summary

#school_summary = school_summary[["School Type", "Students", "School Budget", "Budget per Student", "Average Math Score",
#                                "Average Reading Score", "% Passing Math", "% Passing Reading",
#                                "Overall Pass Rate"]]
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# group by spending per student
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]                       
school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels = group_names)  
groupby_spending = school_data_complete.groupby("spending_bins")
```

```python
# calculate passing percentages per student spending
math_average = groupby_spending["math_score"].mean()
reading_average = groupby_spending["reading_score"].mean()
    
percent_pass_math = school_data_complete[school_data_complete["math_score"] >=70].groupby("spending_bins")["Student ID"].count()/groupby_spending["Student ID"].count()    
percent_pass_read = school_data_complete[school_data_complete["reading_score"] >=70].groupby("spending_bins")["Student ID"].count()/groupby_spending["Student ID"].count()
    
overall_pass = (percent_pass_math + percent_pass_read)/2  
```

```python
scores_by_school_spend = pd.DataFrame({"Average Math Score": math_average,
                                       "Average Reading Score": reading_average,
                                       "% Passing Math": percent_pass_math,
                                       "% Passing Reading": percent_pass_read,
                                       "Overall Passing Percentage": overall_pass})  
scores_by_school_spend
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# calculate math and reading percentages per school size
# group results by school size
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["size_bins"] = pd.cut(school_data_complete["size"], size_bins, labels = group_names)
groupby_size = school_data_complete.groupby("size_bins")
math_average = groupby_size["math_score"].mean()
reading_average = groupby_size["reading_score"].mean()
perc_pass_math = school_data_complete[school_data_complete["math_score"] >=70].groupby("size_bins")["Student ID"].count()/groupby_size["Student ID"].count()
perc_pass_reading = school_data_complete[school_data_complete["reading_score"] >=70].groupby("size_bins")["Student ID"].count()/groupby_size["Student ID"].count()
overall_passing = (perc_pass_math + perc_pass_reading)/2
scores_by_school_size = pd.DataFrame({"Average Math Score": math_average,
                                      "Average Reading Score": reading_average,
                                      "% Passing Reading": perc_pass_reading, 
                                      "% Overall Passing Rate": overall_passing})
scores_by_school_size
```
